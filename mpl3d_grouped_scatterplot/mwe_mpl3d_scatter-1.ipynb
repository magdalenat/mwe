# MWE: 3D scatterplot grouped by id from txt file via pd.DataFrame
keywords: pandas, mpl3d, matplotlib, python, scatterplot3D, subsets

```python
%matplotlib notebook

# reading libraries
# pandas to read and manipulate data from file
import pandas as pd

# necessary for scatterplot 3d in matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# how many segments in dataframe/dataset can be determined as follows
def how_many_segm_in(dataset):
    return len(dataset.color.unique())
```

```python
# read data from file
segmented = pd.read_csv('sample_set.txt', delimiter=',', decimal='.')

# making list of colors from the file (can be set also arbitrary) to set subdataframe color in scatterplot
colors = segmented['color'].unique()
```

```python
subsegm=[]
for segm in range (0, how_many_segm_in(segmented)):
    subsegm.append(segmented[segmented['color'] == colors[segm]])
    
# uncomment below to see sample of set
# subsegm[1].head()
```

```python
# set size of figure
fig = plt.figure(figsize=(6,5))

# set area to plot, with 3d
ax = fig.add_subplot(111, projection='3d')

# add points corresponding to certain subdataframe (each subdataframe in different color)
# Q: how to make it in a loop?
ax0 = ax.scatter(subsegm[0]['x'], subsegm[0]['y'], subsegm[0]['z'], c=colors[0], marker = '.')
ax1 = ax.scatter(subsegm[1]['x'], subsegm[1]['y'], subsegm[1]['z'], c=colors[1], marker = '.')
ax2 = ax.scatter(subsegm[2]['x'], subsegm[2]['y'], subsegm[2]['z'], c=colors[2], marker = '.')
ax3 = ax.scatter(subsegm[3]['x'], subsegm[3]['y'], subsegm[3]['z'], c=colors[3], marker = '.')
ax4 = ax.scatter(subsegm[4]['x'], subsegm[4]['y'], subsegm[4]['z'], c=colors[4], marker = '.')
ax5 = ax.scatter(subsegm[5]['x'], subsegm[5]['y'], subsegm[5]['z'], c=colors[5], marker = '.')
ax6 = ax.scatter(subsegm[6]['x'], subsegm[6]['y'], subsegm[6]['z'], c=colors[6], marker = '.')

# set title
plt.title('Segmented scatterplot 3d')

# setting off axes plot, comment to show axes
plt.axis('off')

plt.show()
```
